Sistem Sederhana Pengelola Nilai Mahasiswa dengan Array

1. Input 10 Nilai

```python
import numpy as np


print("Nilai Mahasiswa: ")
nilaimahasiswa = np.array([89,90,80,55,79,98,95,91,85,60])
nil = np.array(nilaimahasiswa)
print(nil)
```

2. Nilai Tertinggi dna Terendah

```python
nilaitertinggi = np.max(nil)
print("Nilai tertinggi:", nilaitertinggi)


nilaiterendah = np.min(nil)
print("Nilai tertinggi:", nilaiterendah)
```

3. Menghitung Rata-Rata

```python
ratarata = np.mean(nil)
print("Rata-rata:", ratarata)
```

4. Menghitung Jumlah Mahasiswa yang Lulus (>=60)

```python

n = np.array (nil)

lulus = [n for n in nil if n >= 60]
tidak_lulus = [n for n in nil if n < 60]
Jlulus = len(lulus)
Jtidaklulus = len(tidak_lulus)
print(f"Jumlah Lulus   : {Jlulus}")
print(f"Tidak Lulus    : {Jtidaklulus}")
```

5. Grafik Nilai Tertinggi dan Terendah

```python
import matplotlib.pyplot as plt

labels = ["Tertinggi", "Terendah", "Rata-rata"]
values = [nilaitertinggi, nilaiterendah, ratarata]
colors = ['#406093', '#FFF799', '#91D06C']

fig, ax1 = plt.subplots()
ax1.bar(labels, values, color=colors)
ax1.set_title("Grafik Nilai Mahasiswa")
ax1.set_ylabel("Nilai")
plt.show()
```

6. Grafik data Kelulusan

```python
import matplotlib.pyplot as plt


labels = ["Lulus", "Tidak Lulus"]
sizes = Jlulus, Jtidaklulus
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=180)
plt.title("Grafik Data Kelulusan Mahasiswa")
plt.axis('equal')
plt.show()
```

7. Analisis Konpleksitas Setiap Operasi



*   Input Data: $O(n)$ karena melakukan perulangan sebanyak $n$ mahasiswa.
*   Mencari Max/Min: $O(n)$ karena program memeriksa setiap elemen dalam array satu kali.


*   Menghitung Rata-rata: $O(n)$ untuk menjumlahkan semua elemen (sum).
*   Filter Kelulusan: $O(n)$ karena menggunakan list comprehension untuk mengecek setiap nilai.
*   Visualisasi: $O(n)$ untuk memproses data menjadi grafik.


*   Kesimpulan: Secara keseluruhan, algoritma ini berjalan dalam waktu linear atau $O(n)$.
